```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams.update({'font.size': 16})
plt.rcParams['font.family'] = 'Times New Roman'  # 设置字体为Times New Roman

# 1. 准备数据（从示例图中提取的数值）
c = np.array([0,2,4,6,8])  # X轴参数c
mrr = np.array([0.0337,0.0346,0.0355,0.0338,0.0333])      # 左侧Y轴：柱状图MRR@5
ndcg = np.array([0.0392,0.0403,0.0411,0.0394,0.0388])     # 右侧Y轴：折线图NDCG@5

# 2. 创建画布与轴
fig, ax = plt.subplots(figsize=(5, 3))  # 调整尺寸与示例图匹配
ax2 = ax.twinx()  # 创建右侧Y轴

text_color = '#000000'  # 颜色
bar_color = '#1f77b4'  # 蓝色柱状图
line_color = '#d62728'  # 红色折线
# 3. 绘制柱状图（左侧Y轴：MRR@5）
bars = ax.bar(
    x=c, 
    height=mrr, 
    width=1.3,  # 柱状图宽度（适配c的间隔）
    color=bar_color, 
    alpha=0.8, 
    # edgecolor='black',  # 柱状图边框
    linewidth=0.8,
    label='MRR@5'
)

# 4. 绘制折线图（右侧Y轴：NDCG@5）
line, = ax2.plot(
    c, 
    ndcg, 
    color=line_color, 
    marker='o',  # 圆点标记
    markersize=6,
    markerfacecolor=line_color,  # 标记填充色
    # markeredgecolor='black',     # 标记边框
    markeredgewidth=1,
    linewidth=1.2,
    linestyle='-',
    label='NDCG@5'
)

# 5. 设置坐标轴样式
# X轴
#ax.set_xlabel('L', fontsize=13, color=text_color)
ax.set_xticks(c)  # X轴刻度与c对应

# 左侧Y轴（MRR@5）
ax.set_ylabel('MRR@5', fontsize=16, color=text_color)
ax.set_ylim(0.032, 0.036)  # 匹配示例图Y轴范围
ax.tick_params(axis='y', labelcolor=text_color)

# 右侧Y轴（NDCG@5）
ax2.set_ylabel('NDCG@5', fontsize=16, color=text_color)
ax2.set_ylim(0.037, 0.042)  # 匹配示例图Y轴范围
ax2.tick_params(axis='y', labelcolor=text_color)

# 6. 添加网格线（与示例图一致）
ax.grid(True, alpha=0.5, linestyle='--', linewidth=0.5, which='both')
ax2.grid(False)
# 7. 添加组合图例
handles = [bars, line]
labels = [h.get_label() for h in handles]
ax.legend(handles, labels, loc='upper right', fontsize=10, frameon=True, framealpha=0.8, edgecolor='gray')

# 8. 调整布局并显示
plt.tight_layout()
plt.savefig('exp_fmoe_beauty.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
