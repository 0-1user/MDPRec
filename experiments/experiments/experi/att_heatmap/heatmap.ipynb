```python
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
import torch
```

```python
self_att = torch.load('normatt_p.npy')
plin_att = torch.load('normatt.npy')
```

```python
self_att = self_att.cpu().numpy()
plin_att = plin_att.cpu().numpy() 
```

```python


dataset1 = self_att[0, :, :]
dataset2 = plin_att[0, :, :]


vmin = min(np.min(dataset1), np.min(dataset2))
vmax = max(np.max(dataset1), np.max(dataset2))
norm = colors.Normalize(vmin=vmin, vmax=vmax)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))


heatmap1 = ax1.pcolormesh(dataset1, cmap='cividis', norm=norm)
heatmap2 = ax2.pcolormesh(dataset2, cmap='cividis', norm=norm)


ax1.set_title("(a) NormAtt without M", loc='center', y=-0.25)
ax2.set_title("(b) NormAtt", loc='center', y=-0.25)


cbar = fig.colorbar(heatmap1, ax=[ax1, ax2], location='right', shrink=1, pad=-0.45)

plt.tight_layout()
plt.savefig('heatmap.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
