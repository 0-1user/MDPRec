```python
import matplotlib.pyplot as plt
import numpy as np

# 数据准备
categories = ['Beauty', 'LastFM']
methods = ['Ours', 'Coarse Only', 'Fine Only', 'HourOfDay', 'DayOfWeek', 'SeasonOfYear']
colors = ['#446364', '#5D877A', '#6EACA7', '#6E9FAC', '#B5D2D9', '#E6D3D0']

# 每个类别下各方法的NDCG@5值
data_Beauty = [0.0424, 0.0304, 0.0388, 0.0413, 0.0415, 0.0403]
data_LastFM = [0.0303, 0.0220, 0.0268, 0.0263, 0.0255, 0.0264]

# 创建包含两个子图的图形
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# 设置x轴位置
x_pos = np.arange(len(methods))
width = 0.6  # 柱子的宽度

# 绘制Beauty类别的柱状图
bars_toys = axes[0].bar(x_pos, data_Beauty, width, color=colors, edgecolor='black', linewidth=0.8)
#axes[0].set_title('Beauty - NDCG@5', fontsize=12, fontweight='bold', pad=10)
axes[0].set_xlabel('Beauty', fontsize=13)
axes[0].set_ylabel('NDCG@5', fontsize=13)
axes[0].set_xticks([])
#axes[0].set_xticklabels(methods, rotation=45, ha='right', fontsize=9)
axes[0].set_ylim(0.02, 0.043)
axes[0].grid(True, axis='y', alpha=0.3, linestyle='--')

# 绘制LastFM类别的柱状图
bars_beauty = axes[1].bar(x_pos, data_LastFM, width, color=colors, edgecolor='black', linewidth=0.8)
#axes[1].set_title('LastFM - NDCG@5', fontsize=12, fontweight='bold', pad=10)
axes[1].set_xlabel('LastFM', fontsize=13)
axes[1].set_ylabel('NDCG@5', fontsize=13)
axes[1].set_xticks([])
#axes[1].set_xticklabels(methods, rotation=45, ha='right', fontsize=9)
axes[1].set_ylim(0.015, 0.031)
axes[1].grid(True, axis='y', alpha=0.3, linestyle='--')

# 创建统一的图例
legend_elements = []
for i, (method, color) in enumerate(zip(methods, colors)):
    legend_elements.append(plt.Rectangle((0, 0), 1, 1, facecolor=color, edgecolor='black', label=method))


# 将图例放在图形顶部
fig.legend(handles=legend_elements, 
           labels=methods, 
           #title='Method',
           title_fontsize=10,
           fontsize=9,
           loc='upper center', 
           bbox_to_anchor=(0.5, 1.0), 
           ncol=len(methods),
           frameon=True,
           fancybox=True,
           shadow=False)

# 调整布局
#plt.suptitle('NDCG@5', fontsize=14, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.9])  # 为顶部的图例留出空间

# 显示图形


# 可选：保存图形
plt.savefig('time_embedding.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
